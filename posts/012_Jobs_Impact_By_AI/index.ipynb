## Will an LLM Take My Job?

One of the big questions about how AI will effect the world is in what jobs AI will be able to replace. Naturally with all technolgoical development - people are scared about what will happen to their jobs. I want to explore this idea a bit via the tasks that AI will be able to do.

We can do this by examing the **ICAREX Generative AI Impact Matrix** as listed in this Medium Article (https://medium.com/@icarex/icarex-generative-ai-impact-matrix-defining-and-applying-the-strategic-tool-in-business-context-2c47ee7ab88e).

Here we can see the 4 catagories of the matrix:

**Rabbit (Easy for AI, Easy for Humans)** ‚Äî These are the low-hanging fruits, tasks that both AI and humans can perform easily. This is where we‚Äôll likely see the quickest adoption of AI - maybe jobs like data entry, basic customer service chatbots, or scheduling tools. This is largely a good thing freeing up human time for more creative or strategic endeavors - but is likely going to shakeup back office work that is currently being offshored.

**Ant (Easy for AI, Difficult for Humans)** ‚Äî This is where AI truly shines. Tasks that are cognitively or physically demanding for humans but manageable for AI - things like language translation, data pattern recognition, or predictive analytics - are in this catagory. Investing in these areas often leads to disproportionate gains, making them smart medium-term bets for innovation and productivity. Like ants, these applications work tirelessly.

**Eagle (Difficult for AI, Difficult for Humans)** ‚Äî These are the moonshots. Tasks that are tough for both humans and machines ‚Äî such as breakthrough scientific research, complicated medical practices, high-level strategic decision-making, or autonomous systems in unpredictable environments. Progress here is slower but potentially game-changing.

**Hamster (Difficult for AI, Easy for Humans)** ‚Äî These tasks might seem deceptively attractive for automation, but they can become traps. These are things humans do with ease ‚Äî like casual conversation, nuanced customer service, or crafting compelling stories ‚Äî but which remain elusive and resource-intensive for AI to replicate well.


I think that we are going to see the "Rabbits" sprint ahead - potentially destroying other markets where customer service and data entry roles create millions of jobs. The trap is going to be when companies and business leaders attempt to automate the "Eagles" and the "Hamsters". This is where we see jobs that should be being done by humans get automated away in a non-ideal way. I think that software engineering could potentially fall into this "Eagle catagory.
