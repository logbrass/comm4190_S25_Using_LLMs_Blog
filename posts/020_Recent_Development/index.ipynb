## Word Hunt - GamePigeon

Game Pigeon is a popular imessage game for Apple devices that lets friends send games via text. There are a couple of different game options like sports games (like basketball, darts, and archery), classic board games (such as Connect 4, Checkers, and Mancala), and word games. It's like a mini arcade with friends

Word Hunt is one of the most popular games. Players need to find words in a grid of letters. Here are the rules:

    1. The game presents a 4x4 grid of random letters (16 letters total).
    2. Players must form as many words as possible by connecting adjacent letters in the grid. Word Hunt (Game Pigeon) Solver You can move horizontally, vertically, or diagonally from one letter to the next in any direction.
    3.Scoring is based on word length rather than individual letter values
    4. Whoever has the most points at the end of the timer wins.

Having played (and won) the game hundreds of times - I was curious how ChatGPT would perform with this task. In order to explore this I will provide chat GPT with a game board after explaining the rules, then I will run the same board through a Word Hunt solver - we will see which output can score the most points.

GPT did not perform nearly as well as the optimal answer. I should note that this is likely a computational limit then one based on the actual model. Essentially GPT seemed to give up after outputting a certain number of words - not wanting to check others words as GPT was essentially brute forcing words slowly. It is a good example of a task suited for a program rather than an LLM.
